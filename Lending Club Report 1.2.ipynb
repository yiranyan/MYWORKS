# 1 Introduction

In this project, we divide the project to five parts. They are introduction, data cleaning for visualization, visualization about the company profile, user portrait, good and bad loans performance, prediction, and the last one is our conclusion.  
  
In recent years, the rapid development of P2P in China is accompanied by many problems. Many platforms went bankrupt and ran away with money, leaving investors and borrowers with huge losses.
  
So, It is necessary to know about related loan platforms.
  
P2P is a kind of Internet finance (ITFIN), with the help of e-commerce professional network platform to help investors and borrowers to establish a lending relationship and complete relevant transaction procedures.  
In short, it acts as an intermediary platform, using its own risk control and information integrate ability, to screen borrowers with good credit, connect them with investors, and make profits by collecting service fees.
  
Through the investigation, we found that compared with foreign countries, domestic P2P industry has a variety of problems. Therefore, we want to study how to develop in the P2P industry through some well-developed platforms. After filtering, we chose Lending Club.
Lending Club founded in San Francisco in 2006. It is the world's largest online financial platform for matching investors and borrowers currently.

There are three main reasons for the good development of Lending Club:
  
The first one is the business model. It only serves as an information platform and does not involve any asset-side. Provide information to investors to select investment projects, and the risks should be borne by the investors themselves. Based on the information of the borrowers, to evaluate and issue the corresponding low interest rate. Improve borrowing efficiency and reduce borrowing costs.
  
The second point is it has the great FICO and grade standards. FICO credit score is a personal credit rating method developed by the US consumer credit rating agency. Lending club based on the information provided by the borrower, then score and screen quality customers with the score of six hundred and sixty or above, and combine with internal grade ratings (there are seven grades, including A to G, each grade has 5 sub-grades) to reduce the risk of default.
  
The last one is interest rate setting. The setting of fixed interest rate, each grade has a benchmark rate, according to the loan cycle and other conditions to adjust a little interest rate. Provide the lowest interest rate for all types of borrowers to reduce the pressure.
  
Among these risks faced by P2P lending enterprises, the most important risk is the personal credit risk. With this dataset, we want to focus on analyzing the company profile, user portrait and predict default loan. Then help the platform to solve the problem of information asymmetry, provides certain reference opinions for investors and reduce the risk of bad debts.



# 2 Data Cleaning

The raw data has 145 features, but not all of them are useful. Definitions of the columns are given in the Lending Club "Data Dictionary" available here.

We will drop the features with the following criteria:

More than 50% of the data is missing


```python
# Define a function to visulize the features with missing values, and % of total values, & datatype
def missing_values_table(loan_data):
     # Total missing values
    mis_val = loan_data.isnull().sum()
    # Percentage of missing values
    mis_val_percent = 100 * loan_data.isnull().sum() / len(loan_data)
    mis_val_type = loan_data.dtypes
    # Make a table with the results
    mis_val_table = pd.concat([mis_val, mis_val_percent, mis_val_type], axis=1)
        
     # Rename the columns
    mis_val_table_ren_columns = mis_val_table.rename(columns = {0 : 'Missing Values', 1 : '% of Total Values', 2: 'type'})
        
    # Sort the table by percentage of missing descending
    mis_val_table_ren_columns = mis_val_table_ren_columns[ mis_val_table_ren_columns.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False).round(1)
        
    # Print some summary information
    print ("Your selected dataframe has " + str(loan_data.shape[1]) + " columns.\n" "There are " + str(mis_val_table_ren_columns.shape[0]) + " columns that have missing values.")
        
    # Return the dataframe with missing information
    return mis_val_table_ren_columns
```

```python
missing_values_table(loan_data)
```

From this table, we can see that there is a huge gap between features missing some data (around 15% and below) and those missing lots of data (up to 100%). So, we chose 50% as the cutoff. All the missing values exceeded 50%, we deleted them.

```python
def missing_values_table(loan_data):
     # Total missing values
    mis_val = loan_data.isnull().sum()
    # Percentage of missing values
    mis_val_percent = 100 * loan_data.isnull().sum() / len(loan_data)
    mis_val_type = loan_data.dtypes
    # Make a table with the results
    mis_val_table = pd.concat([mis_val, mis_val_percent, mis_val_type], axis=1)
        
     # Rename the columns
    mis_val_table_ren_columns = mis_val_table.rename(columns = {0 : 'Missing Values', 1 : '% of Total Values', 2: 'type'})
        
    # Sort the table by percentage of missing descending
    mis_val_table_ren_columns = mis_val_table_ren_columns[ mis_val_table_ren_columns.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False).round(1)
        
    # Print some summary information
    print ("Your selected dataframe has " + str(loan_data.shape[1]) + " columns.\n" "There are " + str(mis_val_table_ren_columns.shape[0]) + " columns that have missing values.")
        
    # Return the dataframe with missing information
    return mis_val_table_ren_columns
```

```python
# plot the AUROC scores on the training dataset. 
auroc_means = [grid_sgdlr.best_score_, grid_rf.best_score_,grid_knn2.best_score_]
auroc_res = pd.DataFrame({"AUROC":auroc_means,"Algorithm":["SGD Logistic Regression",
"RandomForest","KNeighboors"]})

g = sns.barplot("AUROC","Algorithm", data = auroc_res, palette="Set3",orient = "h")
g.set_xlabel("ROC_AUC score")
g = g.set_title("Cross validation AUROC scores")
auroc_res
```

```python
def evaluation(X_train, X_test, Y_train, Y_test, Y_train_pred, Y_train_prob, Y_pred, Y_prob):
    print("--- ROC AUC ---")
    print("Training Set:", roc_auc_score(Y_train, Y_train_prob))
    print("Test Set:", roc_auc_score(Y_test, Y_prob))
    
    print("\n--- Accuracy ---")
    print("Training Set:", accuracy_score(Y_train, Y_train_pred))
    print("Test Set:", accuracy_score(Y_test, Y_pred))

    tn, fp, fn, tp = confusion_matrix(Y_test, Y_pred).ravel()
    print("\n--- Confusion Matrix ---")
    print("True Positive:", tp)
    print("False Negative:", fn)
    print("True Negative:", tn)
    print("False Positive:", fp)

    print("\n--- Precision ---")
    print("Training Set:", precision_score(Y_train, Y_train_pred))
    print("Test Set:", precision_score(Y_test, Y_pred))

    print("\n--- Recall ---")
    print("Training Set:", recall_score(Y_train, Y_train_pred))
    print("Test Set:", recall_score(Y_test, Y_pred))

    print("\n--- F1 Score ---")
    print("Training Set:", f1_score(Y_train, Y_train_pred))
    print("Test Set:", f1_score(Y_test, Y_pred))
      
def plot_ROC(X_test, Y_test, Y_prob):
    
    #Y_prob = model.predict_proba(X_test)[:,1]
    fpr, tpr, thresh = roc_curve(Y_test, Y_prob, pos_label=1)
    roc_auc = roc_auc_score(Y_test, Y_prob)
     # These are the points at threshold = 0.1~0.5
    x1 = fpr[(thresh <= 0.5) & (thresh >= 0.1)] 
    x2 = tpr[(thresh <= 0.5) & (thresh >= 0.1)]

    fig = plt.figure()
    plt.plot(fpr, tpr, color='r', lw=2)
    plt.plot([0, 1], [0, 1], color='b', lw=2, linestyle='--')
    plt.plot(x1, x2, color='k', lw=3, label='threshold = 0.1 ~ 0.5')
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve (Area = {:.2f})'.format(roc_auc))
    plt.legend(loc="lower right")
    plt.show()
```

```python
print('===== Logistic Regression =====')
evaluation(X_train, X_test, y_train, y_test, y_train_pred_sgdlr, y_train_prob_sgdlr, y_pred_sgdlr, y_prob_sgdlr)
plot_ROC(X_test, y_test, y_prob_sgdlr)
```
